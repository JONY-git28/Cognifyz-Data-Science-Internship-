**Identify the color that represents the highest average rating among different price ranges**

```python
# Calculate average rating for each price range
average_ratings = []
for pr in df['Price range']:
    avg_rating = np.mean(df['Aggregate rating'][df['Price range'] == pr])
    average_ratings.append(avg_rating)

# Find the maximum average rating and corresponding color
max_avg_rating = max(average_ratings)
max_index = average_ratings.index(max_avg_rating)
color_for_max_rating = df['Rating color'][max_index]

print(f"The color that represents the highest average rating among different price ranges is: {color_for_max_rating}")
```

**Encoding categorical variables like 'Has Table Booking' and 'Has Online Delivery' and Scaling Numerical Features**

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
preprocessor = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), ['Cuisines','Has Table booking', 'Has Online delivery']),
        ('scaler', StandardScaler(), ['Average Cost for two', 'Price range', 'Votes'])
    ],
    remainder='passthrough'
)

X = preprocessor.fit_transform(X)
```

```python
print(X.shape)
```

```python
print(X[:10])
```

# **Level 3**

# **Task 1 : Predictive Modeling**

**Split the dataset into training and testing sets**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
print(X_train)
```

**Building and evaluating various regression models**

**Building Linear Regression Model**

```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Create a linear regression model
lr_model = LinearRegression()

# Train the model with the training data
lr_model.fit(X_train, Y_train)

Y_rating_lr = lr_model.predict(X_test)
print(Y_rating_lr)
```

**Evaluating Linear Regresssion Model**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print(f"Model: Linear Regression")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(Y_test, Y_rating_lr)}")
print(f"Mean Squared Error (MSE): {mean_squared_error(Y_test, Y_rating_lr)}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(Y_test,Y_rating_lr))}")
print(f"R-squared (R2) Score: {r2_score(Y_test, Y_rating_lr)}\n")
```

**Building Random Forest Regressor Model**

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor()

rf_model.fit(X_train, Y_train)

Y_predrf = rf_model.predict(X_test)
print(Y_predrf)
```

**Evaluating Random Forest Regressor Model**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print(f"Model:RandomForestRegressor")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(Y_test, Y_predrf)}")
print(f"Mean Squared Error (MSE): {mean_squared_error(Y_test, Y_predrf)}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(Y_test,Y_predrf))}")
print(f"R-squared (R2) Score: {r2_score(Y_test, Y_predrf)}\n")
```

**Building Decision Tree  Regressor Model**

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor

dtr_model = DecisionTreeRegressor()

dtr_model.fit(X_train, Y_train)

Y_preddtr = dtr_model.predict(X_test)
print(Y_preddtr)
```

**Evaluating Decision Tree Regressor Model**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print(f"Model: DecisionTreeRegressor")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(Y_test, Y_preddtr)}")
print(f"Mean Squared Error (MSE): {mean_squared_error(Y_test, Y_preddtr)}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(Y_test,Y_preddtr))}")
print(f"R-squared (R2) Score: {r2_score(Y_test, Y_preddtr)}\n")
```

**Building Gradient Booster Regressor Model**

```python
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor()

gb_model.fit(X_train, Y_train)

Y_predgb = gb_model.predict(X_test)
print(Y_predgb)
```

**Evaluating Gradient Booster Regressor Model**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
print(f"Model: GradientBoostingRegressor")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(Y_test, Y_predgb)}")
print(f"Mean Squared Error (MSE): {mean_squared_error(Y_test, Y_predgb)}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(Y_test,Y_predgb))}")
print(f"R-squared (R2) Score: {r2_score(Y_test, Y_predgb)}\n")
```
